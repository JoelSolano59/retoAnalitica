```python
# Import libraries
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
insurance = pd.read_csv('https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv')
insurance.head(3)
```

```python
insurance = insurance.replace(regex=['yes'], value=1)
insurance = insurance.replace(regex=['no'], value=0)
insurance = insurance.replace(regex=['female'], value=0)
insurance = insurance.replace(regex=['male'], value=1)
insurance.head(3)
```

```python
sns.histplot(data = insurance, x = 'charges', hue = 'sex', bins = 10, kde = True)
sns
```

```python
sns.histplot(data = insurance, x = 'charges', hue = 'children', bins = 10, kde = True)
sns
```

```python
insurance_corr = insurance.corr()
sns.heatmap(data = insurance_corr, vmin = 0, vmax = 1, cmap = 'RdBu', annot = True, square = True)
sns
```

```python
fig = plt.figure(figsize=(9,6))
sns.boxplot(data=insurance, x='charges', y='region')
sns
```

```python
insurance = insurance.replace(regex=['southwest'], value=0)
insurance = insurance.replace(regex=['southeast'], value=1)
insurance = insurance.replace(regex=['northwest'], value=2)
insurance = insurance.replace(regex=['northeast'], value=3)
insurance["region"].plot(kind="hist")
```
